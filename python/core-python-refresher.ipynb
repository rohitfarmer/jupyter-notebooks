# Getting Back to Core Python

A refresher on Python fundamentals: data types, structures, control flow, functions, and file handling.

## 1. Variables and Data Types

```
x = 10
x = 'hello'
x
```

## 2. Strings

```
s = 'python'
s[0], s[-1], s[1:4]
```

## 3. Lists

```
nums = [10, 20, 30]
nums.append(40)
nums
```

## 4. Tuples

```
point = (10, 20)
point
```

## 5. Dictionaries

```
user = {'name': 'Ada', 'age': 32}
user['name']
```

## 6. Sets

```
tags = {'python', 'coding', 'python'}
tags
```

## 7. Conditions

```
x = 10
if x > 5:
    result = 'big'
else:
    result = 'small'
result
```

## 8. Loops

```
for i in range(3):
    print(i)
```

## 9. Functions

```
def add(a, b):
    return a + b
add(2, 3)
```

## 10. Nested Data Access

```
data = {'users': [{'name': 'Ada'}, {'name': 'Linus'}]}
data['users'][0]['name']
```

## 11. File Handling

```
with open('example.txt', 'w') as f:
    f.write('Hello file')

with open('example.txt') as f:
    f.read()
```

## 12. Summary
Python programs typically:
1. Read input
2. Process data
3. Write output
